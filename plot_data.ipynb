```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open('NIS_radar.txt') as f:
   rad_data = f.readlines()

rad_data = [float(x.strip()) for x in rad_data]
```

```python
plt.plot(range(1,len(rad_data)),rad_data[1:],range(1,len(rad_data)),7.8*np.ones(len(rad_data)-1))
plt.show()
```

```python
with open('NIS_laser.txt') as f:
   las_data = f.readlines()

las_data = [float(x.strip()) for x in las_data]
```

```python
plt.plot(range(1,len(las_data)),las_data[1:],range(1,len(las_data)),6.0*np.ones(len(las_data)-1))
plt.show()
```
